- Considere a "Matriz de Eisenhower" para interpretar o gráfico "Reasons for Leaving".
- Na Matriz de Eisenhower, as tarefas são classificadas com base em sua urgência e importância, onde as tarefas urgentes e importantes ocupam o topo à direita, enquanto as tarefas menos importantes estão na parte inferior à esquerda.

- Aplicando esse conceito ao gráfico "Reasons for Leaving", podemos considerar as razões de saída da seguinte maneira:

###Urgente e Importante:
- As razões que estão no canto superior direito do gráfico são as mais relevantes e críticas. Essas são as questões que devem receber atenção imediata, como as oportunidades de carreira e a falta de reconhecimento, que têm altos pesos e, portanto, são urgentes e importantes para resolver.

###Importante, mas Menos Urgente:
- As razões que estão no canto superior esquerdo são importantes, mas podem não ser tão urgentes. Estas podem incluir o tipo de trabalho, que é significativo, mas pode não exigir ação imediata.

###Urgente, mas Menos Importante:
- As razões que estão no canto inferior direito são urgentes, mas menos importantes. Estas podem ser questões que devem ser tratadas, mas que têm um impacto menor, como preocupações com carga de trabalho.

###Menos Urgente e Menos Importante:
- As razões no canto inferior esquerdo são menos urgentes e menos importantes. Essas questões podem não ser uma prioridade imediata, como mudanças de carreira ou realocações, que têm um peso relativamente baixo.

## Gráfico de colunas empilhadas.
- Vantagem:
 - Facilidade de leitura: Eles são fáceis de entender e interpretar, mesmo para pessoas que não têm experiência em análise de dados, devido à sua natureza visual e intuitiva.
- Desvantagem:
 - Complexidade com muitas categorias: Quando há muitas categorias empilhadas, o gráfico pode ficar confuso e difícil de ler. É importante evitar o excesso de categorias para manter a clareza.

## Gráfico de barras lado a lado:
- Vantagem:
 - Comparação direta: Os gráficos de barras lado a lado facilitam a comparação direta de valores entre categorias ou grupos, já que as barras não estão empilhadas umas sobre as outras.
- Desvantagem:
 - Complexidade com muitas categorias: Quando há muitas categorias ou grupos para comparar, os gráficos de barras lado a lado podem se tornar confusos e desordenados, dificultando a análise.


image.png

## Gráfico de barras horizontais em cores com uma faixa média de referência.
- Vantagem:
 - Espaço para rótulos: Como as barras são horizontais, há mais espaço para rótulos ao longo do eixo vertical, tornando-os ideais quando você tem rótulos de categoria longos.
- Desvantagem:
 - Menos espaço para rótulos de valores: Embora haja mais espaço para rótulos de categoria, há menos espaço para rótulos de valores ao longo do eixo horizontal. Isso pode dificultar a leitura dos valores exatos.
